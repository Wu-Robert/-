```python
full_seat=(df["場均"]/df["容納人數"])*100
fig=plt.figure(figsize=(25,10))
plt.plot(df["球隊"],full_seat,'o',)
```

```python
op=pd.read_csv("OffensePower.csv")
fire=pd.DataFrame(df,columns=["球隊"])
fire["得分"]=(op["得分"]+op["全壘打"]*1.2+op["打擊率"]+op["上壘率"]+op["長打率"]*2)/100
xfire=fire.sort_index(by="得分",ascending=False)
xfire
```

```python
dp=pd.read_csv("Defense.csv")
de=pd.DataFrame(df,columns=["球隊"])
de["防禦"]=dp["防禦率"]
xde=de.sort_index(by="防禦")
xde
```

```python
mul=pd.DataFrame(df,columns=["球隊"])
mul["綜合指標"]=fire["得分"]*10/(de["防禦"]*de["防禦"])
xmul=mul.sort_index(by="綜合指標",ascending=False)
xmul
```

```python
matplotlib.rc("xtick",labelsize=10)
matplotlib.rc("ytick",labelsize=20)
plt.rc("axes",titlesize=20)
plt.rc("axes",labelsize=20)
fig=plt.figure(figsize=(12,5))
fig=plt.figure(figsize=(15,5))
rects1=plt.bar(np.arange(30),
               fire["得分"],0.25,
               alpha=0.4,color="steelblue",
               label="攻擊力")
rects2=plt.bar(np.arange(30)+0.25,
               (10-de["防禦"])*0.5,0.25,
               alpha=0.4,color="orange",
               label="防禦力")
rects3=plt.bar(np.arange(30)+0.5,
               mul["綜合指標"],0.25,
               alpha=0.4,color="green",
               label="綜合指標")
plt.xlabel("球隊")
plt.ylabel("指數")
plt.title("球隊指標")
plt.xticks(np.arange(30)+0.35,df["球隊"])
plt.legend()
plt.tight_layout()
plt.show()
```
